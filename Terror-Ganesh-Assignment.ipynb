- Q.7)label all the incidents where the number killed was more than 5 as severe. 

```python
terror['severe']= ['severe' if kill>5 else float('nan') for kill in terror.nkill]
```

- Q.8)write a function to label an incident that was both successful and suicidal

```python
def suc(terror):
    return (np.where(((terror.suicide==1) & (terror.success==1)),'suicide success','suicide failure'))
terror['suc-suc']=suc(terror)
```
